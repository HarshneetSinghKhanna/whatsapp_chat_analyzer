```python
import re
import pandas as pd
```

```python
f = open('Suhani_chat.txt', 'r',encoding = 'utf-8')
```

```python
data = f.read()
```

```python
print((data))
```

```python


pattern = r'\d{2}/\d{2}/\d{2}, \d{2}:\d{2}:\d{2}\]?\s*'
```

```python
messages = re.split(pattern,data)[1:]
messages
```

```python
dates = re.findall(pattern,data)
dates
```

```python
import pandas as pd

df = pd.DataFrame({'user_message': messages, 'message_date': dates})


df['message_date'] = df['message_date'].str.replace(']', '', regex=True).str.strip()


df['message_date'] = pd.to_datetime(df['message_date'], format='%d/%m/%y, %H:%M:%S')


df.rename(columns={'message_date': 'date'}, inplace=True)
df.head()
```

```python
df.shape
```

```python
# separating users and messages
users = []
messages = []

for message in df['user_message']:
    entry = message.split(':', 1)  # Splitting at the first occurrence of ':'
    if len(entry) == 2:
        users.append(entry[0])  # User name
        messages.append(entry[1])  # Actual message
    else:
        users.append("group_notification")  # Handling system messages
        messages.append(entry[0])  # The entire message is stored

df['user'] = users
df['message'] = messages

df.drop(columns=['user_message'], inplace=True)

df.head()
```

```python
df['year'] = df['date'].dt.year
```

```python
df.head()
```

```python
df['month'] = df['date'].dt.month_name()
df['day'] = df['date'].dt.day_name()
df['hour'] = df['date'].dt.hour
df['minute'] = df['date'].dt.minute
```
